```python
crime_columns = ['Murder', 'Assault on women', 'Kidnapping and Abduction', 
                 'Dacoity', 'Robbery', 'Arson', 'Hurt', 
                 'Prevention of atrocities (POA) Act', 
                 'Protection of Civil Rights (PCR) Act', 
                 'Other Crimes Against SCs']
```

```python
df['Total Crimes'] = df[crime_columns].sum(axis=1)
```

```python
#drop null values from columns
df.dropna(axis=1)
```

```python
#converting data types
df['Year']=pd.to_numeric(df['Year'],errors='coerce')
df['Murder']=pd.to_numeric(df['Murder'],errors='coerce')
df['Assault on women']=pd.to_numeric(df['Assault on women'],errors='coerce')
df['Kidnapping and Abduction']=pd.to_numeric(df['Kidnapping and Abduction'],errors='coerce')
df['Dacoity']=pd.to_numeric(df['Dacoity'],errors='coerce')
df['Robbery']=pd.to_numeric(df['Robbery'],errors='coerce')
df['Arson']=pd.to_numeric(df['Arson'],errors='coerce')
df['Hurt']=pd.to_numeric(df['Hurt'],errors='coerce')
df['Prevention of atrocities (POA) Act']=pd.to_numeric(df['Prevention of atrocities (POA) Act'],errors='coerce')
df['Protection of Civil Rights (PCR) Act']=pd.to_numeric(df['Protection of Civil Rights (PCR) Act'],errors='coerce')
df['Other Crimes Against SCs']=pd.to_numeric(df['Other Crimes Against SCs'],errors='coerce')
```

```python
#Encoding categorical variables
df['STATE/UT']=df['STATE/UT'].astype('category')
df['DISTRICT']=df['DISTRICT'].astype('category')
```

```python
df.to_csv("clean_crime_district.csv",index=False)
```

```python
df1 = pd.read_csv("clean_crime_district.csv")
```

```python
# Group by district and sum murders
murder_df = df1.groupby('DISTRICT')['Murder'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
bars = plt.bar(murder_df.index, murder_df.values, color='firebrick', edgecolor='black')

# Add value labels on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=10)

# Style
plt.title('ðŸ”ª Top 10 Districts with Highest Murders', fontsize=16, fontweight='bold')
plt.ylabel('Total Murders')
plt.xlabel('District')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Group and sort states by total crimes
state_crime_totals = df1.groupby('STATE/UT')['Total Crimes'].sum().sort_values(ascending=True)  # horizontal chart works better ascending

# Plot the horizontal bar chart
plt.figure(figsize=(15, 10))
bars = plt.barh(state_crime_totals.index, state_crime_totals.values, color='darkorchid')

# Add value labels
for bar in bars:
    plt.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width()):,}', va='center', fontsize=9, color='black')

# Styling
plt.title("Total Crime Counts Across States/UTs", fontsize=16, fontweight='bold')
plt.xlabel("Total Crimes", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.gca().set_facecolor('#f9f9f9')
plt.show()
```

```python
crime_types = ['Murder', 'Assault on women', 'Kidnapping and Abduction', 'Robbery', 'Hurt']
df1[crime_types] = df1[crime_types].apply(pd.to_numeric, errors='coerce')

# Group by Year and sum crimes
trend_df = df1.groupby('Year')[crime_types].sum()

# Plotting the trends
plt.figure(figsize=(14, 7))
for crime in crime_types:
    plt.plot(trend_df.index, trend_df[crime], marker='o', label=crime)

plt.title('Trend Analysis of Common Crime Types Over the Years', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.legend(title='Crime Types')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
```

```python
# Group by year and sum POA and PCR crimes
act_trend = df1.groupby('Year')[['Prevention of atrocities (POA) Act', 'Protection of Civil Rights (PCR) Act']].sum()

# Plotting stacked bar chart
plt.figure(figsize=(12, 6))
act_trend.plot(kind='bar', stacked=True, color=['darkred', 'navy'])

plt.title('Crimes Under POA and PCR Acts Over the Years (Stacked)', fontsize=15, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend(title='Crime Type')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Group by District and calculate the average total crimes (across all years)
district_avg_crime = df1.groupby('DISTRICT')['Total Crimes'].mean()

# Plot histogram
plt.figure(figsize=(12, 6))
plt.hist(district_avg_crime, bins=20, color='skyblue', edgecolor='black')

plt.title('Histogram: Average Total Crimes per District', fontsize=16, fontweight='bold')
plt.xlabel('Average Total Crimes')
plt.ylabel('Number of Districts')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Group by state and sum murders
state_murder_df = df1.groupby('STATE/UT')['Murder'].sum().sort_values(ascending=False).head(10)

# Plot pie chart
plt.figure(figsize=(7, 7))
colors = plt.cm.Set3.colors  # Use a colorful palette
plt.pie(state_murder_df.values, labels=state_murder_df.index, autopct='%1.1f%%',
        startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title('ðŸ©¸ Top 10 States/UTs with Highest Murders', fontsize=16, fontweight='bold')
plt.axis('equal')  # Keep the pie chart circular
plt.tight_layout()
plt.show()
```

```python
# Group by Year and State/UT
state_year_crimes = df1.groupby(['Year', 'STATE/UT'])['Total Crimes'].sum().unstack().fillna(0)

# Select top 10 states with highest total crimes
top_states = state_year_crimes.sum().sort_values(ascending=False).head(10).index
heatmap_data = state_year_crimes[top_states].T  # Transpose for heatmap

# Plot heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlOrRd', linewidths=0.3, linecolor='gray')

plt.title('ðŸ”¥ Heatmap: State-wise Total Crimes Over the Years', fontsize=16, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('State/UT')
plt.tight_layout()
plt.show()
```
